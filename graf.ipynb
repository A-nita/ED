```python

from matplotlib import pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline
plt.rcParams['figure.figsize'] = (11,7)


```

```python
def media(vetor):
    soma = 0
    for i in vetor:
        soma = soma + i
    return soma/5
        
```

```python
y_dez = [30, 33, 27, 27, 35]
y_cem = [2312,2553,2519,2587,2429]
y_mil = [249814,249193,246515,244718,250588]
y_dez_mil = [24340601,24299764,24364023,24461081,23999326]

a = media(y_dez)
b = media(y_cem)
c = media(y_mil)
d = media(y_dez_mil)

e = np.array([a, b, c,6216331, d]) # Cria um array do Numpy para ter acesso a novas funcionalidades (e mais desempenho)

dez = [10, 10, 10, 10, 10]
cem = [100, 100, 100, 100, 100]
mil = [1000, 1000, 1000, 1000, 1000]
dez_mil = [10000, 10000, 10000, 10000, 10000]

entrada = np.array([10,100,1000, 5000, 10000])


```

```python


# 300 represents number of points to make between T.min and T.max
xnew = np.linspace(entrada.min(), entrada.max(), 300)  

power_smooth = make_interp_spline(entrada, e)
power_smooth = power_smooth(xnew)
```

```python
with plt.style.context('dark_background'):
    plt.plot(dez, y_dez, 'o', cem, y_cem, 'o', mil, y_mil, 'o', dez_mil, y_dez_mil, 'o', xnew,power_smooth)

plt.ylabel('Comparacoes')
plt.xlabel('Entradas')

plt.show()
```
